# 1. Tuples

-   Basics
-   Ordered collection, written with ().

```python
my_tuple = (10, 20, 30, "GenAI")
```

### Characteristics

-   Immutable → cannot change values once created.
-   Can store mixed data types.
-   Allow duplicates.
-   Indexed: my_tuple[0] → 10.

## Operations

```python
# Access
print(my_tuple[2])
```

```python
# Slicing:
print(my_tuple[1:3])
```

```python
# Concatenation:
new_tuple = my_tuple + ("Python",)
```

```python
new_tuple
```

## Methods

-   count() → occurrences.
-   index() → position.

```python
nums = (1, 2, 2, 3)
```

```python
print(nums.count(2))
```

```python
print(nums.index(3))
```

## Packing & Unpacking

```python
student = ("Rock", 30, "AI")
```

```python
name, age, course = student
```

```python
print(name, age, course)
```

# 2. Strings

**Basics**

Sequence of characters enclosed in quotes.

```python
text = "Hello GenAI"
```

```python
# Common Operations
print("Hello" + " World")
```

```python
# Repetition:
print("Hi! " * 3)
```

```python
# Indexing & Slicing:
print(text[0])
```

```python
print(text[0:5])
```

## Useful Methods

-   `upper()`, `lower()`, `title()`
-   `strip()` → remove spaces
-   `replace(old, new)`
-   `split()` → break into list
-   `join()` → combine list into string

```python
msg = " python basics "
```

```python
print(msg.strip())
```

```python
print(msg.upper())
```

```python
print(msg.replace("python", "GenAI"))
```

```python
sentence = "Python is fun"
```

```python
print("Python" in sentence)
```
