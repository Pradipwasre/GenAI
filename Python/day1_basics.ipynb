# Python Basics

## Variables

-   Store values for later use.

```python
name = "Rock"
age = 30
```

```python
# let print the variables

print(name)
print(age)
```

Data Types

-   **int** → numbers (10, -5)
-   **float** → decimals (3.14)
-   **str** → text ("Hello")
-   **bool** → True/False

```python
user_name = input("Enter your name: ")
```

```python
print("Welcome,", user_name)
```

## Operators

-   Arithmetic: `+` `-` `*` `/` `%` `//` **`
-   Comparison: `==`  `!=` `>` `<` `>=` `<=`
-   Logical: `and` or `not`

```python
x = 5
y = 3
print(x + y) 
print(x > y) 
```
